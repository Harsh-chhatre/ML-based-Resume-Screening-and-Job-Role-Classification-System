```python
import pandas as pd
import numpy as np
import nltk
import string
```

```python
dataset= pd.read_csv("UpdatedResumeDataSet.csv")
```

```python
dataset.head(4)
```

```python
nltk.download("stopwords")
nltk.download("punkt_tab")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
def clean_text(text):
    text= text.lower()
    text= text.translate(str.maketrans('','',string.punctuation))
    words= word_tokenize(text)
    stop_words= set(stopwords.words('english'))
    word= [word for word in words if word not in stop_words]
    stemmer= PorterStemmer()
    words= [stemmer.stem(word) for word in words]
    return " ".join(words)
```

```python
dataset["clnd_resume"] = dataset["Resume"].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf= TfidfVectorizer(max_features=2000)
x= tfidf.fit_transform(dataset["clnd_resume"]).toarray()
```

```python
y= dataset["Category"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb= MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
print(mnb.score(x_test,y_test)*100,mnb.score(x_train,y_train)*100)
```

```python
pred= mnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print("\nAccuracy:", accuracy_score(y_test, pred))
print("\nClassification Report:\n", classification_report(y_test, pred))
```
